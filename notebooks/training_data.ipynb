# Training dataset for image classification

We use CIFAR-10 dataset as training dataset for our image classification model (https://www.cs.toronto.edu/~kriz/cifar.html). In this code, we will check how the data looks like. Notice that we have 10 classes, and 32 by 32 image and 3 color channels. It is important information when we build models and detection algorithm.

## Setup

```python
from tensorflow.keras.datasets import cifar10
import numpy as np
from matplotlib import pyplot as plt
```

## Hyperparameter

```python
label_name = np.array(['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck'])
NUM_IMAGES = 10
```

## Data

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
print("Number of training dataset: {}".format(x_train.shape[0]))
print("Number of test dataset: {}".format(x_test.shape[0]))
```

```python
for i in range(NUM_IMAGES):
    plt.imshow(x_train[i])
    plt.title(label_name[int(y_train[i])])
    plt.show()
```

```python
print("Shape of each image: {}".format(x_train[0].shape))
print("Height: {}".format(x_train[0].shape[0]))
print("Width: {}".format(x_train[0].shape[1]))
print("Color channel: {}".format(x_train[0].shape[2]))
plt.imshow(x_train[0])
plt.show()
```
