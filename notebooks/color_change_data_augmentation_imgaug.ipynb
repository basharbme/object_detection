# Data augmentation to change color with imgaug

```python
## setup

import numpy as np
import matplotlib.pyplot as plt

import imgaug as ia
import imgaug.augmenters as iaa
```

```python
## sample image

images = np.array(
    [ia.quokka(size=(64, 64)) for _ in range(32)],
    dtype=np.uint8
)

print("shape:", images.shape)
plt.imshow(images[0,:,:,:])
plt.show()
```

```python
seq = iaa.Sequential([
    iaa.MultiplyHue((0.5, 1.5))
])

images_aug = seq(images = images)

plt.figure(figsize = (16, 8))
for i in range(4*8):
    plt.subplot(4, 8, i+1)
    plt.imshow(images_aug[i])
    plt.axis('off')
plt.show()
```

```python
seq = iaa.Sequential([
    iaa.MultiplyHueAndSaturation(mul_hue = (0.5, 1.5))
])

images_aug = seq(images = images)

plt.figure(figsize = (16, 8))
for i in range(4*8):
    plt.subplot(4, 8, i+1)
    plt.imshow(images_aug[i])
    plt.axis('off')
plt.show()
```

```python
plt.figure(figsize = (16, 8))

for i in range(4*8):
    plt.subplot(4, 8, i+1)
    plt.imshow(images_aug[i])
    plt.axis('off')

plt.show()
```
