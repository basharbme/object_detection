# Data Augmentation

Read P.138 5.2.5 Using data augmentation in "Deep Learning with Python" by Francois book. Model 2 looks at a green part in an image and incorrectly predicts it as a frog. This code intends to improve this point.

## Setup

```python
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
from matplotlib import pyplot as plt
```

## Hyperparameter

```python
label_name = np.array(['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck'])
FROG_IDX = int(np.where(label_name == 'frog')[0])
NUM_IMAGES = 10
NUM_SHOW = 20
```

## Data

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
print(x_train.shape)
print(x_test.shape)
print("The number of frog images:", np.sum(y_train == FROG_IDX))
```

```python
FROG_IDXS = np.where(y_train == FROG_IDX)[0]
x_train_frog = x_train[FROG_IDXS]

for i in range(NUM_SHOW):
    plt.imshow(x_train_frog[i])
    plt.show()
```
